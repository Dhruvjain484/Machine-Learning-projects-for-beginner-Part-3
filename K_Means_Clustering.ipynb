##Importing the libraries


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##Importing the dataset


```
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```
x = dataset.iloc[:, 2:].values
X= x.copy()
dataset.corr()     #Checking the correlation between differnt features
```

##Finding optimum number of clusters

```
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters = i, init = 'k-means++')  #k-means++ is an algorithm to avoid the k-means random initialization trap
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.xlabel('range')
plt.ylabel('wcss')
plt.title('the elbow method');
```

##Training the K-means cluster model on the dataset

```
# Not always using the K-means clustering gives the good result

from sklearn.cluster import KMeans
kmeans_1 = KMeans(n_clusters = 5, init= 'k-means++', random_state = 0)
y_means = kmeans_1.fit_predict(x)
y_means
```

#Visualising the model

#Annual Income vs spending score

```
# before Clustering
plt.scatter(X[:,1], X[:, 2])
plt.xlabel('Annual Income')
plt.ylabel('spending score')
plt.title("Before Clustering")
plt.show();


# After Clustering
plt.scatter(x[y_means==0, 1], x[y_means==0, 2], color = 'r')
plt.scatter(x[y_means==1, 1], x[y_means==1, 2], color = 'b')
plt.scatter(x[y_means==2, 1], x[y_means==2, 2], color = 'g')
plt.scatter(x[y_means==3, 1], x[y_means==3, 2], color = 'y')
plt.scatter(x[y_means==4, 1], x[y_means==4, 2], color = 'c')
plt.scatter(kmeans_1.cluster_centers_[:,1], kmeans_1.cluster_centers_[:, 2], color = 'black', s = 200)
plt.xlabel('Annual Income')
plt.ylabel('spending score')
plt.title('After Clustering');
```
