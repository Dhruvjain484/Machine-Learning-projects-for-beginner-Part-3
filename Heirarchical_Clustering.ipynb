##Importing the libraries


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##Importing the dataset


```
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```
x = dataset.iloc[:, 2:].values
X = x.copy()
dataset.corr()     #Checking the correlation between differnt features
```

##Finding optimum number of clusters by using Dendrogram

```
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('customers')
plt.ylabel('euclidean distance')
```

##Training the Heirarchical cluster model on the dataset

```
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters = 5)  #Try n_cluster = 3
y_cluster = clustering.fit_predict(x)
y_cluster
```

#Visualising the model

#Annual Income vs spending score

```
# before Clustering
plt.scatter(X[:,1], X[:, 2])
plt.xlabel('Annual Income')
plt.ylabel('spending score')
plt.title("Before Clustering")
plt.show();


# After Clustering
plt.scatter(x[y_cluster==0, 1], x[y_cluster==0, 2], color = 'r')
plt.scatter(x[y_cluster==1, 1], x[y_cluster==1, 2], color = 'b')
plt.scatter(x[y_cluster==2, 1], x[y_cluster==2, 2], color = 'g')
plt.scatter(x[y_cluster==3, 1], x[y_cluster==3, 2], color = 'y')
plt.scatter(x[y_cluster==4, 1], x[y_cluster==4, 2], color = 'c')
# plt.scatter(clustering.cluster_centers_[:,1], clustering.cluster_centers_[:, 2], color = 'black', s = 200)
plt.xlabel('Annual Income')
plt.ylabel('spending score')
plt.title('After Clustering');
```

#Age vs spending score

```
# Before Clustering
plt.scatter(X[:,0], X[:,2])
plt.xlabel('age')
plt.ylabel('spending score')
plt.title('Before Clustering')
plt.show();


# After Clustering
plt.scatter(x[y_cluster==0, 0], x[y_cluster==0, 2], color = 'r')
plt.scatter(x[y_cluster==1, 0], x[y_cluster==1, 2], color = 'b')
plt.scatter(x[y_cluster==2, 0], x[y_cluster==2, 2], color = 'g')
plt.scatter(x[y_cluster==3, 0], x[y_cluster==3, 2], color = 'y')
plt.scatter(x[y_cluster==4, 0], x[y_cluster==4, 2], color = 'c')
# plt.scatter(clustering.cluster_centers_[:,0], clustering.cluster_centers_[:, 2], color = 'black', s = 200)

plt.xlabel('age')
plt.ylabel('spending score')
plt.title('After Clustering');
```
